```python
# Definimos la función que nos dibujará los graficos de histograma y boxplot para las variables numéricas

#https://stackoverflow.com/questions/33381330/histogram-with-boxplot-above-in-python

def histogram_boxplot(data, xlabel = None, title = None, figsize=(10,8), bins = None):
    """ Boxplot and histogram combined
    data: 1-d data array
    xlabel: xlabel 
    title: title
    font_scale: the scale of the font (default 2)
    figsize: size of fig (default (9,8))
    bins: number of bins (default None / auto)

    example use: histogram_boxplot(np.random.rand(100), bins = 20, title="Fancy plot")
    """
    
    f2, (ax_box2, ax_hist2) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)}, figsize=figsize)
    sns.boxplot(data, ax=ax_box2)
    sns.distplot(data, ax=ax_hist2, bins=bins, kde=False) if bins else sns.distplot(data, ax=ax_hist2,kde=False)
    if xlabel: ax_hist2.set(xlabel=xlabel)
    if title: ax_box2.set(title=title)
    ax_box2.set(xlabel='')

    plt.show()
```

```python
#Resumen estadístico variables categóricas


# Definimos la función que nos dibujará los graficos de barras para las variables categoricas con su frecuencia

#https://stackoverflow.com/questions/33179122/seaborn-countplot-with-frequencies

def plot_freq_catvar(var_cat, datos=None) -> None:
    
    dataplot = travel_raw if datos is None else datos

    #ncount = data[var_cat].notnull().sum()
    ncount = len(dataplot)

    plt.figure(figsize=(12,8))

    num_cat=len(dataplot[var_cat].value_counts().index)

    if num_cat > 10:
        ax = sns.countplot(x=var_cat,data=dataplot, order=dataplot[var_cat].value_counts().iloc[:10].index) 
    else:
        ax = sns.countplot(x=var_cat,data=dataplot, order=dataplot[var_cat].value_counts().index)

    ax.set_xticklabels(ax.get_xticklabels(),rotation=45, horizontalalignment='right')
    
    # Make twin axis
    ax2=ax.twinx()

    # Switch so count axis is on right, frequency on left
    ax2.yaxis.tick_left()
    ax.yaxis.tick_right()

    # Also switch the labels over
    ax.yaxis.set_label_position('right')
    ax2.yaxis.set_label_position('left')

    ax2.set_ylabel('Frequency [%]')

    for p in ax.patches:
        x=p.get_bbox().get_points()[:,0]
        y=p.get_bbox().get_points()[1,1]
        ax.annotate('{:.1f}%'.format(100.*y/ncount), (x.mean(), y), 
                ha='center', va='bottom') # set the alignment of the text

    # Use a LinearLocator to ensure the correct number of ticks
    ax.yaxis.set_major_locator(ticker.LinearLocator(11))

    # Fix the frequency range to 0-100
    ax2.set_ylim(0,100)
    ax.set_ylim(0,ncount+1000)

    # And use a MultipleLocator to ensure a tick spacing of 10
    ax2.yaxis.set_major_locator(ticker.MultipleLocator(10))

    # Need to turn the grid on ax2 off, otherwise the gridlines end up on top of the bars
    ax2.grid(None)
    
    plt.show()
```

## Resample sobre Train data

### SMOTE - Random Undersample

```python
# Oversample with SMOTE and random undersample for imbalanced dataset

from collections import Counter
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline

# summarize class distribution
counter = Counter(y)
print(counter)
# define pipeline
over = SMOTE(sampling_strategy=0.05, random_state=42)
under = RandomUnderSampler(sampling_strategy=0.5, random_state=42)
steps = [('o', over), ('u', under)]
pipeline = Pipeline(steps=steps)
# transform the dataset
X_sm, y_sm = pipeline.fit_resample(X_train_bp, y_train)
X_sm_pca, y_sm_pca = pipeline.fit_resample(X_train_bp_pca, y_train)
# summarize the new class distribution
counter = Counter(y_sm)
print(counter)
```

### SMOTETomek

```python
from imblearn.combine import SMOTETomek
from imblearn.under_sampling import TomekLinks


counter = Counter(y)
print(counter)

# define sampling strategy
resample = SMOTETomek(tomek=TomekLinks(sampling_strategy='majority'),random_state=42)
# define pipeline
pipeline = Pipeline(steps=[('r', resample)])

# transform the dataset
X_smt, y_smt = pipeline.fit_resample(X_train_bp, y_train)
X_smt_pca, y_smt_pca = pipeline.fit_resample(X_train_bp_pca, y_train)
# summarize the new class distribution
counter = Counter(y_smt)
print(counter)
```
