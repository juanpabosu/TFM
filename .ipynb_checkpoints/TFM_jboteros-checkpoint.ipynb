```python
# Definimos la función que nos dibujará los graficos de histograma y boxplot para las variables numéricas

#https://stackoverflow.com/questions/33381330/histogram-with-boxplot-above-in-python

def histogram_boxplot(data, xlabel = None, title = None, figsize=(10,8), bins = None):
    """ Boxplot and histogram combined
    data: 1-d data array
    xlabel: xlabel 
    title: title
    font_scale: the scale of the font (default 2)
    figsize: size of fig (default (9,8))
    bins: number of bins (default None / auto)

    example use: histogram_boxplot(np.random.rand(100), bins = 20, title="Fancy plot")
    """
    
    f2, (ax_box2, ax_hist2) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)}, figsize=figsize)
    sns.boxplot(data, ax=ax_box2)
    sns.distplot(data, ax=ax_hist2, bins=bins, kde=False) if bins else sns.distplot(data, ax=ax_hist2,kde=False)
    if xlabel: ax_hist2.set(xlabel=xlabel)
    if title: ax_box2.set(title=title)
    ax_box2.set(xlabel='')

    plt.show()
```

```python
#Resumen estadístico variables categóricas


# Definimos la función que nos dibujará los graficos de barras para las variables categoricas con su frecuencia

#https://stackoverflow.com/questions/33179122/seaborn-countplot-with-frequencies

def plot_freq_catvar(var_cat, datos=None) -> None:
    
    dataplot = travel_raw if datos is None else datos

    #ncount = data[var_cat].notnull().sum()
    ncount = len(dataplot)

    plt.figure(figsize=(12,8))

    num_cat=len(dataplot[var_cat].value_counts().index)

    if num_cat > 10:
        ax = sns.countplot(x=var_cat,data=dataplot, order=dataplot[var_cat].value_counts().iloc[:10].index) 
    else:
        ax = sns.countplot(x=var_cat,data=dataplot, order=dataplot[var_cat].value_counts().index)

    ax.set_xticklabels(ax.get_xticklabels(),rotation=45, horizontalalignment='right')
    
    # Make twin axis
    ax2=ax.twinx()

    # Switch so count axis is on right, frequency on left
    ax2.yaxis.tick_left()
    ax.yaxis.tick_right()

    # Also switch the labels over
    ax.yaxis.set_label_position('right')
    ax2.yaxis.set_label_position('left')

    ax2.set_ylabel('Frequency [%]')

    for p in ax.patches:
        x=p.get_bbox().get_points()[:,0]
        y=p.get_bbox().get_points()[1,1]
        ax.annotate('{:.1f}%'.format(100.*y/ncount), (x.mean(), y), 
                ha='center', va='bottom') # set the alignment of the text

    # Use a LinearLocator to ensure the correct number of ticks
    ax.yaxis.set_major_locator(ticker.LinearLocator(11))

    # Fix the frequency range to 0-100
    ax2.set_ylim(0,100)
    ax.set_ylim(0,ncount+1000)

    # And use a MultipleLocator to ensure a tick spacing of 10
    ax2.yaxis.set_major_locator(ticker.MultipleLocator(10))

    # Need to turn the grid on ax2 off, otherwise the gridlines end up on top of the bars
    ax2.grid(None)
    
    plt.savefig(var_cat+'.pdf')
```

```python
#Imputar valores en Duration menores o iguales a cero con mean_duration_by_dest
for key in mean_duration_by_dest.index.values:
    travel_clean.loc[travel_clean['duration'] <= 0 & (travel_clean['destination']==key) , 'duration']=mean_duration_by_dest[key]
```

```python
#Analizar outliers en variable Age
travel_clean[(travel_clean["age"]) > 100 ]
```

```python
#Analizar outliers en variable Age
travel_clean[(travel_clean["age"]) == 0 ]
```

```python
#Para las edades supueriores a 100 años o iguales a cero, se imputaran valores con la mediana de la variable Age

age_median=np.median(travel_clean['age'])
travel_clean.loc[travel_clean["age"] >100,'age'] = age_median
travel_clean.loc[travel_clean["age"] == 0,'age'] = age_median
```

```python
travel_clean.describe()
```

```python
# Cambiar tipo de dato a category para las variables categóricas
# category	Finite list of text values

cols_to_exclude = ['age', 'duration', 'net_sales', 'commision']
for col in travel_clean.columns:
    if travel_clean[col].nunique() < 200 and col not in cols_to_exclude:
        travel_clean[col] = travel_clean[col].astype('category')
```

```python
travel_clean.info()
```

```python
travel_clean.head(10)
```

```python
# One hot encoding y lower case variables categoticas predictoras
columns=['agency','agency_type','product_name','destination']

for column in columns:
    travel_clean[column]=travel_clean[column].str.lower()
    one_hot = pd.get_dummies(travel_clean[column])
    # Drop first column in one_hot
    one_hot=one_hot.drop(one_hot.columns[[0]], axis=1)
    # Drop column as it is now encoded
    travel_clean = travel_clean.drop(column,axis = 1)
    # Join the encoded df
    travel_clean = travel_clean.join(one_hot)    
```

```python
# Label encoding en target: claim
le = preprocessing.LabelEncoder()
travel_clean['claim_label'] = le.fit_transform(travel_clean.claim.values)

travel_clean = travel_clean.drop('claim',axis = 1)
```

```python
# Estandarización en variables numéricas predictoras
from sklearn.preprocessing import StandardScaler

#columns=['duration','net_sales','commision','age','art']
columns=travel_clean.columns[travel_clean.columns != 'claim_label']
scaler = StandardScaler()
travel_clean[columns]=scaler.fit_transform(travel_clean[columns])
```

```python
travel_clean.columns = travel_clean.columns.str.replace(' ','_')
travel_clean.columns = travel_clean.columns.str.replace(',','')

travel_clean.reset_index(drop=True,inplace=True)
```

## Boruta

Feature Selection is one of the key step in machine learning. It reduces the computation time and also may help in reducing over-fitting. There are several ways to select features like RFE, Boruta and Genetic Algorithms.
Today I am going to demonstrate how to use Boruta for feature selection in python. Boruta by default uses random forest although it works with other algorithms like LightGBM, XGBoost etc. It selects the features that actually have relationship with outcome variable. The algorithm reshuffles the data to create shadow features. It eliminates the features that have significantly worst importance than shadow ones. There are few limitations in the approach. It is required to remove multicollinearity and missing values from data before running Boruta.

https://medium.com/@vikeshsingh37/feature-selection-using-boruta-e5a3605ff161

```python
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy as bp

X = travel_clean.loc[ : , travel_clean.columns != 'claim_label'].values 
y = travel_clean['claim_label'].values
feature_names=travel_clean.columns[travel_clean.columns != 'claim_label']

rf_model = RandomForestClassifier(n_jobs=-1, class_weight='balanced', max_depth=5)
feat_selector = bp(rf_model,n_estimators = 'auto', verbose= 0,max_iter= 100)
feat_selector.fit(X, y)

selected_features = [feature_names[i] for i, x in enumerate(feat_selector.support_) if x]
print(selected_features)
```

```python
X_filtered = feat_selector.transform(X)
X_filtered
```

# Aplicación de modelos

## Entrenamiento y test

Antes de aplicar algún modelo, tenemos que separar los datos entre los conjuntos de train y test. Siempre trabajaremos sobre el conjunto de train y evaluaremos los resultados en el conjunto de test.

```python
from sklearn.model_selection import train_test_split
    
X_train, X_test, y_train, y_test = train_test_split(X_filtered, y, test_size=0.33, random_state=42)
```

```python
# Oversample with SMOTE and random undersample for imbalanced dataset

from collections import Counter
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline

# summarize class distribution
counter = Counter(y)
print(counter)
# define pipeline
over = SMOTE(sampling_strategy=0.1)
under = RandomUnderSampler(sampling_strategy=0.5)
steps = [('o', over), ('u', under)]
pipeline = Pipeline(steps=steps)
# transform the dataset
X_sm, y_sm = pipeline.fit_resample(X_train, y_train)
# summarize the new class distribution
counter = Counter(y_sm)
print(counter)
```
